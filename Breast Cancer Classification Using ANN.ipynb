```python
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
plt.figure(figsize=(25,23))
sns.heatmap(x.corr(),annot=True,fmt='.2f')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scale_data_x = sc.fit_transform(x)
x=pd.DataFrame(scale_data_x,columns=x.columns)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print('x_train:',len(x_train))
print('x_test:',len(x_test))
print('y_train:',len(y_train))
print('y_test:',len(y_test))
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(50,3),learning_rate_init=0.1,max_iter=100,random_state=2)
model.fit(x_train,y_train)
predit_test = model.predict(x_test)
predit_train = model.predict(x_train)
predit_proba = model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Train accuracy: ",accuracy_score(y_train,predit_train))
print("Test accuracy: ",accuracy_score(y_test,predit_test))
```

```python
print(confusion_matrix(y_test,predit_test))
```

```python
print(confusion_matrix(y_train,predit_train))
```

```python
print(classification_report(y_test,predit_test))
```
